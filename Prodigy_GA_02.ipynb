```python
import torch
from diffusers import StableDiffusionPipeline
from IPython.display import display
```

```python
model_id = "runwayml/stable-diffusion-v1-5"

pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16
)

pipe = pipe.to("cuda")
```

```python
import torch
print(torch.cuda.is_available())
print(torch.cuda.get_device_name(0) if torch.cuda.is_available() else "No GPU")
```

```python
prompt = "A futuristic city at sunset with flying cars, ultra realistic, 4k"

image = pipe(prompt).images[0]

display(image)
```

```python
prompts = [
    "A majestic lion sitting on a mountain during sunrise, ultra realistic",
    "A cyberpunk street at night with neon lights and rain",
    "A beautiful village surrounded by snowy mountains, photorealistic"
]

for prompt in prompts:
    print(f"Prompt: {prompt}")
    image = pipe(prompt).images[0]
    display(image)
```
